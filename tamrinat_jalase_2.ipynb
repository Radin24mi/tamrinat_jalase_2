```python
#جدول درصد فروش دو شرکت در شش ماه
import matplotlib.pyplot as plt

month = ["december","february","march","may","april","january"]

company_A = [20,40,50,10,15,45]
company_B = [80,69,58,77,50,45]

plt.plot(month , company_A , color = "red" , label = ("company asghar"))
plt.plot(month , company_B , color = "blue" , linestyle = "dashed" , label = ("company akbar"))

plt.scatter(month , company_A , color = "yellow" , label = ("company asghar"),s=250)
plt.scatter(month , company_B , color = "yellow" , linestyle = "dashed" , label = ("company akbar"),s=250)


plt.grid(True)
plt.xlabel("month")
plt.ylabel("sell")
```
